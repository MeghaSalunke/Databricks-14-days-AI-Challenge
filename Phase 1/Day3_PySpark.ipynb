```python
# Read the November data
nov = spark.read \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .csv("/Volumes/workspace/ecommerce/ecommerce_data/2019-Nov.csv")

display(nov.limit(10))
```

```python
#Read the October data
oct = spark.read \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .csv("/Volumes/workspace/ecommerce/ecommerce_data/2019-Oct.csv")

display(oct.limit(10))
```

```python
#Inner join
inner_join = nov.join(
    oct,
    nov.user_id == oct.user_id,
    "inner"
).limit(5)
display(inner_join)
```

```python
#Left join
left_join = nov.join(
    oct,
    nov.user_id == oct.user_id,
    "left"
).limit(5)
display(left_join)
```

```python
#Right Join
right_join = nov.join(
    oct,
    nov.user_id == oct.user_id,
    "right"
).limit(5)
display(right_join)
```

```python
#Full Join
full_join = nov.join(
    oct,
    nov.user_id == oct.user_id,
    "outer"
).limit(5)
display(full_join)
```
